# 701 串列[List]數組轉換
請撰寫一程式，輸入數個整數並儲存至串列中，以輸入-9999為結束點（串列中不包含-9999），再將此串列轉換成數組，最後顯示該數組以及其長度（Length）、最大值（Max）、最小值（Min）、總和（Sum）。

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)
[Note]從串列List[] 轉換成數組Tuple()

```python
import pyinputplus as pyip
import random

class InputSection:
    def __init__(self):
        self.list_input=None
        self.tuple_input=None
    def input_method(self):
        while True:
            n=pyip.inputInt("一數組Tuple(), 有多少個元素? 請輸入個數 n: ")
            max_value=pyip.inputInt("請輸入一數組元素的上限值(最大值): ")
            min_value=pyip.inputInt("請輸入一數組元素的下限值(最小值): ")
            if min_value < max_value:
               self.list_input=[random.randint(min_value,max_value) for _ in range(n)]
               self.tuple_input=tuple(self.list_input)               
               return self.tuple_input                 
            else:    
               print("上限值(最大值)要大於下限值(最小值), 請重新輸入")
               continue

class CalculationSection:
    def __init__(self,tuple_input:tuple[int]):
        self.tuple_input=tuple_input
        self.input_min=None
        self.input_max=None
        self.input_sum=None
        self.input_length=None
        
    def calculation_method(self):
        self.input_min=min(self.tuple_input)
        self.input_max=max(self.tuple_input)
        self.input_sum=sum(self.tuple_input)
        self.input_length=len(self.tuple_input)
        
        return self.input_min,self.input_max,self.input_sum, self.input_length 

class DisplaySection:
    def __init__( self,tuple_input:tuple[int], input_min:int,input_max:int,input_sum:int,input_length:int ):
        
        self.tuple_input=tuple_input
        self.input_min=input_min
        self.input_max=input_max
        self.input_sum=input_sum
        self.input_length=input_length
        
    def output_method(self):
        print()
        print(f" 隨機產生的數組為: {self.tuple_input}")
        print(f" 其最大值為: {self.input_max}")
        print(f" 其最小值為: {self.input_min}")
        print(f" 總和為: {self.input_sum}")
        print(f" 長度(length):數組內有{self.input_length}個元素(elements)")        

def main():
    while True:
        try:
            selection:str=pyip.inputYesNo(prompt="Python 701 串列[List]數組轉換:繼續計算請按'y'; 離開程式請按'-9999' ",yesVal="y",noVal="-9999")            
            
            if selection=='y':            
                input_object=InputSection()
                tuple_input=input_object.input_method()
                calculation_object=CalculationSection(tuple_input)
                input_min,input_max,input_sum, input_length=calculation_object.calculation_method()
                display_object=DisplaySection(tuple_input, input_min,input_max,input_sum, input_length)
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

# 702 數組(Tuple)合併排序
請撰寫一程式，輸入並建立兩組數組，各以-9999為結束點（數組中不包含-9999）。將此兩數組合併並從小到大排序之，顯示排序前的數組和排序後的串列

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)
[Note] 建立兩組數組Tuple()的方式 

```python
import pyinputplus as pyip
import random

class InputSection:
    def __init__(self):
        self.tuple_inputa=()
        self.tuple_inputb=()
    
    def input_method(self):
        while True:
            n=pyip.inputInt("隨機產生一數組Tuple 其中的元素個數為? 請輸入個數 n: ",min=1)
            max_value=pyip.inputInt("請輸入一數組的上限值(最大值): ")
            min_value=pyip.inputInt("請輸入一數組的下限值(最小值): ")
            if min_value < max_value:
               inputa=[ random.randint(min_value,max_value) for _ in range (n) ]
               self.tuple_inputa=tuple( map(int, inputa))
               self.tuple_inputb=tuple(map(int,[random.randint(min_value,max_value) for _ in range(n)] ))
               return self.tuple_inputa, self.tuple_inputb                 
            else:    
               print("上限值(最大值)要大於下限值(最小值), 請重新輸入")
               continue

class CalculationSection:
    def __init__(self,tuple_inputa:tuple[int], tuple_inputb:tuple[int] ):
        self.tuple_inputa=tuple_inputa
        self.tuple_inputb=tuple_inputb
        self.tuple_inputab=None
        
    def calculation_method(self):
        self.tuple_inputab=self.tuple_inputa+self.tuple_inputb
        self.sort_inputab=tuple(sorted(self.tuple_inputab))
        return self.tuple_inputab,self.sort_inputab 

class DisplaySection:
    def __init__( self,tuple_inputa:tuple[int], tuple_inputb:tuple[int], tuple_inputab:tuple[int], sort_inputab:tuple[int] ):
        
        self.tuple_inputa=tuple_inputa
        self.tuple_inputb=tuple_inputb
        self.tuple_inputab=tuple_inputab
        self.sort_inputab=sort_inputab
                
    def output_method(self):
        print()
        print(f" 隨機產生的二個數組分別為:\n {self.tuple_inputa} 和 {self.tuple_inputb}\n")
        print(f" 將二個數組合併後, 顯示為:\n {self.tuple_inputab}\n ")
        print(f" 再經過正向排序, 結果為:\n {self.sort_inputab}\n")   

def main():
    while True:
        try:
            selection: str=pyip.inputYesNo(prompt="Python 702 數組合併排序:繼續計算請按'y'; 離開程式請按'-9999' ",yesVal="y",noVal="-9999")            
            
            if selection=='y':            
                input_object=InputSection()
                tuple_inputa, tuple_inputb=input_object.input_method()
                calculation_object=CalculationSection(tuple_inputa, tuple_inputb)
                tuple_inputab,sort_inputab =calculation_object.calculation_method()
                display_object=DisplaySection(tuple_inputa, tuple_inputb,tuple_inputab,sort_inputab )
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

# 702-1 矩陣(Array)合併排序
請撰寫一程式，輸入並建立兩組矩陣，各以-9999為結束點（矩陣中不包含-9999）。將此兩矩陣合併後並從小到大排序之，顯示排序前的矩陣和排序後的矩陣

```python
import pyinputplus as pyip
import random
import numpy as np

class InputSection:
    def __init__(self):
        self.array_inputa=None
        self.array_inputb=None
    
    def input_method(self):
        while True:
            n=pyip.inputInt("隨機產生一矩陣Array 其中的元素個數為? 請輸入個數 n: ",min=1)
            max_value=pyip.inputInt("請輸入一矩陣內元素的上限值(最大值): ")
            min_value=pyip.inputInt("請輸入一矩陣內元素的下限值(最小值): ")
            if min_value < max_value:
               self.array_inputa=np.random.randint(min_value,max_value, size=n)
               self.array_inputb=np.random.randint(min_value,max_value, size=n)
               return self.array_inputa, self.array_inputb                 
            else:    
               print("上限值(最大值)要大於下限值(最小值), 請重新輸入")
               continue

class CalculationSection:
    def __init__(self,array_inputa:np.ndarray[int], array_inputb:np.ndarray[int] ):
        self.array_inputa=array_inputa
        self.array_inputb=array_inputb
        self.horizontal_inputab=None
        self.sort_inputab=None
        
    def calculation_method(self):
        self.horizontal_inputab=np.hstack(( np.array(self.array_inputa), np.array(self.array_inputb)))
        self.sort_inputab=np.sort(self.horizontal_inputab)        
        return self.horizontal_inputab, self.sort_inputab

class DisplaySection:
    def __init__( self,array_inputa:np.ndarray[int], array_inputb:np.ndarray[int], horizontal_inputab:np.ndarray[int], sort_inputab:np.ndarray[int] ):
        
        self.array_inputa=array_inputa
        self.array_inputb=array_inputb        
        self.horizontal_inputab=horizontal_inputab
        self.sort_inputab=sort_inputab
                
    def output_method(self):
        print()
        print(f" 隨機產生的二個矩陣分別為:\n {self.array_inputa} 和 {self.array_inputb}")
        print(f" 將二個矩陣水平(horizontal)合併後, 顯示為:\n {self.horizontal_inputab} ")
        print(f" 再經過正向排序, 結果為:\n {self.sort_inputab}")   

def main():
    while True:
        try:
            selection: str=pyip.inputYesNo(prompt="Python 702 數組合併排序:繼續計算請按'y'; 離開程式請按'-9999' ",yesVal="y",noVal="-9999")            
            
            if selection=='y':            
                input_object=InputSection()
                array_inputa, array_inputb=input_object.input_method()
                calculation_object=CalculationSection(array_inputa, array_inputb)
                horizontal_inputab, sort_inputab =calculation_object.calculation_method()
                display_object=DisplaySection(array_inputa, array_inputb,horizontal_inputab,sort_inputab)
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

[Hint]\
子集合（Subset）和超集合（Superset）的定義：\
1.子集合（Subset）：如果集合 A 的所有元素都在集合 B 中，則稱 A 為 B 的子集合，記作 
A⊆B。\
2. 超集合（Superset）：如果集合 B 包含集合 A 的所有元素，則稱 B 為 A 的超集合，記作
