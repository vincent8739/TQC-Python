# 801 字串索引
請撰寫一程式，要求使用者輸入一字串，顯示該字串每個字元的索引。

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip

class InputSection:
    def __init__(self):
        self.string_input=None
    
    def input_method(self):
        self.string_input=pyip.inputStr("請輸入一字串: ")
        return self.string_input

class CalculationSection:
    def __init__(self,string_input):
        self.string_input=string_input
        self.list_result=[]
        
    def calculation_method(self):
        self.list_result=list( enumerate( self.string_input))
        return self.list_result        

class DisplaySection:
    def __init__(self,list_result:list[tuple[int,str]]):
        self.list_result=list_result
        
    def output_method(self):
        print(f"顯示該字串每個字元的索引為:\n")
        for index, value in self.list_result:
            print(f"{index}:{value}")

def main():
    while True:
        try:
            selection:str=pyip.inputYesNo(prompt="Python 801 字串索引:繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")            
            
            if selection=='y':            
                input_object=InputSection()
                string_input=input_object.input_method()
                calculation_object=CalculationSection(string_input)
                list_result=calculation_object.calculation_method()
                display_object=DisplaySection(list_result)
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

### [Key Note]:
解包（unpacking）是一種方便的技術，用於可迭代(Iterable)的對象（如列表List[]、元組Tuple() 等）中, 提取多個值，並將它們分配給多個變數

# 809 密碼規則
請撰寫一程式，要求使用者輸入一個密碼（字串），檢查此密碼是否符合規則。密碼規則如下：\
　a. 必須至少八個字元。\
　b. 只包含英文字母和數字。\
　c. 至少要有一個大寫英文字母。\
　d. 若符合上述三項規則，程式將顯示檢查結果為【Valid password】，否則顯示【Invalid password】。

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip

class InputSection:
    def __init__(self):
        self.string_input=None
    
    def input_method(self):
        user_input=input("請輸入一個密碼（字串）(包含英文字母和數字)")
        self.string_input=str(user_input)
        return self.string_input

class CalculationSection:
    def __init__(self, string_input:str):
        self.string_input=string_input
        self.check_length=None
        self.check_upper=None
        self.check_alnum=None
            
    def calculation_method(self):
        self.check_length=len(self.string_input)
        self.check_upper=any( char.isupper() for char in self.string_input)
        self.check_alnum=all( char.isalnum() for char in self.string_input)
        return self.check_length, self.check_upper, self.check_alnum

class DisplaySection:
    def __init__(self,string_input:str,check_length:int,check_upper: bool,check_alnum: bool   ):
        self.string_input=string_input
        self.check_length=check_length
        self.check_upper=check_upper
        self.check_alnum=check_alnum
        
    def output_method(self):
        if self.check_length >=5 and self.check_upper and self.check_alnum:
            print()
            print(f"此密碼{self.string_input}是符合規則 【Valid password】! ")
        else:
            print(f"此密碼{self.string_input} 不符合規則 【Invalid password】! ")

def main():
    while True:
        try:
            selection:str=pyip.inputYesNo(prompt="Python 809 密碼規則 :繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")            
            
            if selection=='y':            
                input_object=InputSection()
                string_input=input_object.input_method()
                calculation_object=CalculationSection(string_input)
                check_length,check_upper,check_alnum=calculation_object.calculation_method()
                display_object=DisplaySection(string_input, check_length, check_upper, check_alnum)
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

# 810 最大值與最小值之差
請撰寫一程式，首先要求使用者輸入正整數k（1 <= k <= 100），代表有k筆測試資料。每一筆測試資料是一串數字，每個數字之間以一空白區隔，請找出此串列數字中最大值和最小值之間的差。
提示：差值輸出到小數點後第二位。

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip
import random

class InputSection:
    def __init__(self):
        self.number_list=[]
    
    def input_method(self):
        n=pyip.inputInt("測試資料是一串數字, 請輸入有幾筆數字:",min=1)
        for _ in range (n):
            number=random.randint(1,100)
            self.number_list.append(number)
        return self.number_list

class CalculationSection:
    def __init__(self,number_list:list[int]):
        self.number_list=number_list
        self.number_min=None
        self.number_max=None
        self.number_dif=None
        
    def calculation_method(self):
        self.number_min=min(self.number_list)
        self.number_max=max(self.number_list)
        self.number_dif=self.number_max-self.number_min
        return self.number_min, self.number_max, self.number_dif    

class DisplaySection:
    def __init__( self,number_list:list[int],number_min:int, number_max:int, number_dif:int ):
        self.number_list=number_list
        self.number_min=number_min
        self.number_max=number_max
        self.number_dif=number_dif
        
    def output_method(self):
        print()
        print(f"串列數字為:{' '.join([ str(num) for num in self.number_list ]) } ")
        
        print(f"串列數字的最大值:{self.number_max} 和最小值:{self.number_min} ")
        print(f"其之間的差:{self.number_dif} ")

def main():
    while True:
        try:
            selection:str=pyip.inputYesNo(prompt="Python :繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")            
            
            if selection=='y':
                k=pyip.inputInt("請輸入一正整數k [1<= k <=100] 代表有k筆測試資料: ",min=1,max=100)
                for _ in range (k):             
                    input_object=InputSection()
                    number_list=input_object.input_method()
                    calculation_object=CalculationSection(number_list)
                    number_min, number_max, number_dif=calculation_object.calculation_method()
                    display_object=DisplaySection(number_list, number_min, number_max, number_dif)
                    display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```

### [Key Note]:
>class DisplaySection:        
>>    def output_method(self):\
>>        print(f"串列數字為: {self.number_list}")\
輸出顯示為一 [List] 例如: [37, 15, 53, 55, 11]

1. 遍歷"（Iteration or Traversal）: 是逐一存取列表（List）中的每個元素
> for num in self.number_list 
2. [List] 裡面的元素是int, 再將每個元素(int)轉化成(str)
> [str(num) for num in self.number_list ]
3. 格式化: 刪除 [ , ], 每個字元用空白連接
> '空格'.join(要合併的對象)

完整程式為:
>class DisplaySection:        
>>    def output_method(self):\
>>        print(f"串列數字為: { ' '.join( [str(num) for num in self.number_list]) }")

# 811 猜數字
請撰寫一程式，要求使用者先輸入一個兩位數字．做為要猜的數字。程式將隨機生成兩位數字，並與用戶輸入的數字進行比對。
如果生成的數字與用戶輸入的數字相同，則遊戲結束並顯示成功猜到的次數。

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip
import random

class InputSection:
    def __init__(self):
        self.ssn_key=None
        self.ssn_input=None
    
    def input_method(self):
        self.ssn_key=pyip.inputStr("從0-9 數字, 設定一組二位數 :")
        return self.ssn_key  

class CalculationSection:
    def __init__(self,ssn_key:str):
        self.ssn_key=ssn_key
        self.count=None
        self.ssn_input=None

        
    def calculation_method(self):
        self.count=0
        numbers="0123456789"
        while True:
            self.ssn_input=str( ''.join( random.choices(numbers,k=2)))
            self.count+=1                     
            if self.ssn_input==self.ssn_key:
                return self.count, self.ssn_input
            else:
                continue          

class DisplaySection:
    def __init__(self,count:int, ssn_input:str):
        self.count=count
        self.ssn_input=ssn_input
        
    def output_method(self):
            print()
            print(f"輸入的數字為:{self.ssn_input}, 總共試了{self.count}次")

def main():
    while True:
        try:
            selection:str=pyip.inputYesNo(prompt="Python 811 猜數字 :繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")            
            
            if selection=='y':            
                input_object=InputSection()
                ssn_key=input_object.input_method()
                calculation_object=CalculationSection(ssn_key)
                count,ssn_input=calculation_object.calculation_method()
                display_object=DisplaySection(count, ssn_input)
                display_object.output_method()
            else:
                print("程式結束,謝謝使用! ")
                break        
        
        except pyip.RetryLimitException:
                print("輸入錯誤次數過多，請重新啟動程式。")
                return
        except Exception as e:
                print(f"發生無法預期的錯誤: {e}")
                return

if __name__=="__main__":
     main()
```
