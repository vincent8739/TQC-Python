# 501 訊息顯示
請撰寫一程式，呼叫函式compute()，該函式功能為讓使用者輸入系別（Department）、學號（Student ID）和姓名（Name）並顯示這些訊息。

##### 程式I: 函數式編程(Functional Programming, FP)

```python
import pyinputplus as pyip

def compute(department:str,studentId:int,name:str):
    print("系別:", department)
    print("學號:", studentId)
    print("姓名:", name)
try: 

    department:str=pyip.inputStr("請輸入系別: ")
    studentId:int=pyip.inputInt("請輸入學號: ")
    name:str=pyip.inputStr("請輸入姓名: ")
    compute(department,studentId,name)

except Exception as error:
        print(f"發生未預期的錯誤, {error}")
```

# 502 乘積
請撰寫一程式，將使用者輸入的兩個整數作為參數傳遞給一個名為compute(x, y)的函式，此函式將回傳x和y的乘積。

##### 程式I: 函數式編程(Functional Programming, FP)

```python
import pyinputplus as pyip

def compute(x:int,y:int):
    product=x*y
    return product
try: 
    x=pyip.inputInt ("please input the first number :")
    y=pyip.inputInt ("please input the second number :")
    calculation=compute(x,y)
    print(f"{x}x{y}={calculation}")
except Exception as error:
    print(f"An unexpected error has occurred: {error}")
```

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip

class ProductCalculation:
    def __init__(self,n:int,m:int) -> None:
        self.number1=n
        self.number2=m
        self._product=None
        
    @property
    def product(self):    
        return self._product
           
    def calculation_method(self) -> None:
        self._product=self.number1*self.number2

    def output_method(self) -> None:
        print(f"{self.number1}x{self.number2}={self._product}")

def input_function():
    n=pyip.inputInt("please input the first number:\n")
    m=pyip.inputInt("please input the second number:\n")
    return n, m

def main():

    while True:
        try:
            selection: str=pyip.inputYesNo(prompt="Python 502 乘積, 繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")
            print()

            if selection=='y':
                n,m=input_function()
                product_instance=ProductCalculation(n,m)
                product_instance.calculation_method()
                product_instance.output_method()

            else:
                print("程式結束,謝謝使用! ")
                break

            
        except pyip.RetryLimitException as error:
            print(f"輸入超過限制次數: {error}")
            break
        except ValueError as error:
            print(f"輸入值錯誤: {error}")
            break
        except Exception as error:
            print(f"發生無法預期的錯誤: {error}")
            break

if __name__=="__main__":
    main()
 
```

# 503 連加計算
請撰寫一程式，讓使用者輸入兩個整數，接著呼叫函式compute()，此函式接收兩個參數a、b，並回傳從a連加到b的和。

##### 程式I: 函數式編程(Functional Programming, FP)

```python
import pyinputplus as pyip

def compute(a:int,b:int):
    total=0
    for i in range (a,b+1):
        total+=i
    return total
while True:
    try:
        num1:int=pyip.inputInt("please input the first number: ")
        num2:int=pyip.inputInt("please input the second number: ")
        if num1<num2:
            count:int=compute(a=num1,b=num2)
            print(f"The sum of adding from {num1} to {num2} is {count}")
            break
        else:
            print("Error: The first number must be less than the second number. Please try again")

    except Exception as error:
        print(f"An unexpected error has occurred: {error}")
```

# 505 依參數格式化輸出
請撰寫一程式，將使用者輸入的三個參數，變數名稱分別為a（代表字元character）、x（代表個數）、y（代表列數），作為參數傳遞給一個名為compute()的函式，該函式功能為：一列印出x個a字元，總共印出y列。\
提示：輸出的每一個字元後方有一空格。

##### 程式I: 函數式編程(Functional Programming, FP)

```python
import pyinputplus as pyip

def compute(a:str,x:int,y:int):
    for i in range (y):
        for j in range (x):
            print(a, end='')
        print()
try:
    a=pyip.inputStr("請輸入一個字元: \n")
    x=pyip.inputInt("請輸入個數: \n")
    y=pyip.inputInt("請輸入列數: \n")
    compute(a,x,y)

except Exception as error:
    print(f"發生無法預期的錯誤, {error}")
```

1. print() 的位置：將換行的 print() 放在內層迴圈外面，這樣可以確保每一行的所有字元都會在同一行印出，然後再換行

2. 省略 print( comput(a,x,y) )：因為 compute 函數並不返回任何值，因此不需要在最後印出 print


##### 程式II: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip

class FormatOutput:

    def __init__(self) -> None:
        self.character:str=""
        self.quantity:int=0
        self.column:int=0

    def input_method(self) -> None:
        self.character=pyip.inputStr("請輸入一個字元: \n")
        self.quantity=pyip.inputInt("請輸入個數: \n")
        self.column=pyip.inputInt("請輸入列數: \n")

    def show_table_method(self) -> None:
        for i in range (self.column):
            for j in range (self.quantity):
                print(self.character, end='')
            print()
       
    
def main():
    formatdisplay= FormatOutput()

    try:
        formatdisplay.input_method()
        formatdisplay.show_table_method()
    
    except pyip.RetryLimitException:
            print("輸入錯誤次數過多，請重新啟動程式。")
    except ValueError as e:
            print(f"輸入錯誤: {e}")
    except Exception as e:
            print(f"發生無法預期的錯誤: {e}")

if __name__=="__main__":
     main()
```

# 506 一元二次方程式

請撰寫一程式，將使用者輸入的三個整數（代表一元二次方程式 
𝑎𝑥2+𝑏𝑥+𝑐=0 的三個係數 (a、b、c）作為參數傳遞給一個名為compute()的函式，該函式回傳方程式的解，如無解則輸出【Your equation has no root.】\
提示：\
輸出有順序性\
回傳方程式的解，無須考慮小數點位數

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip
import cmath

class QuadraticEquation:
    def __init__(self) -> None:
        self.a:int=0
        self.b:int=0
        self.c:int=0

    def input_method(self) -> None:
        self.a=pyip.inputInt("請輸入一元二次方程式中的常數 a: \n",min=1)
        self.b=pyip.inputInt("請輸入一元二次方程式中的常數 b: \n")
        self.c=pyip.inputInt("請輸入一元二次方程式中的常數 c: \n")

    def calculation_method(self) -> None:
        d=self.b*self.b-4*self.a*self.c
        if d>0:
             x1=(-self.b + cmath.sqrt(d))/(2*self.a)
             x2=(-self.b - cmath.sqrt(d))/(2*self.a)
             print(f"aX^2+bX+C=0, 方程有兩個不同的實根: X1={x1.real},X2={x2.real} ")
        elif d==0:
             x=(-self.b)/(2*self.a)
             print(f"aX^2+bX+C=0, 方程有一個重根: X={x} ")       
        else:
             x1=(-self.b + cmath.sqrt(d))/(2*self.a)
             x2=(-self.b - cmath.sqrt(d))/(2*self.a)
             print(f"aX^2+bX+C=0, 方程有沒有實根 而是兩個共軛複根: X1={round(x1,ndigits=3)} ,X2={round(x2,ndigits=3)}")

def main():
    root=QuadraticEquation()

    try:
        root.input_method()
        root.calculation_method()
    
    except ZeroDivisionError:
            print("錯誤: 常數 a 不能為零。")

    except pyip.RetryLimitException:
            print("輸入錯誤次數過多，請重新啟動程式。")
 
    except Exception as e:
            print(f"發生無法預期的錯誤: {e}")

if __name__=="__main__":
     main()
```

使用輾轉相除法 (歐幾里得算法 Euclidean algorithm)

例子：a=48, b=18\
•	48/18=12\
•	18/12=6\
•	12/6=0\
•	最大公因數是 6

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip
import math

class GreatestCommonDivisor:
    def __init__(self) -> None:
        self.number1:int=0
        self.number2:int=0
        self.result:int=0
        

    def input_method(self) -> None:
        while True:
            self.number1=pyip.inputInt("請輸入第一個正整數: \n",min=1)
            self.number2=pyip.inputInt("請輸入第二個正整數: \n",min=1)
            if self.number1<=self.number2:
                return
            else:
                print("第二個正整數要大於第一個數, 請重新輸入:")

    def calculation_method(self) -> None:
        self.result=math.gcd(self.number1,self.number2)

    def output_method(self) -> None:
        print(f"兩個正整數:{self.number1}和{self.number2} 其最大公因數是{self.result}")
    
def main():
    GCDcalculation=GreatestCommonDivisor()

    try:
        GCDcalculation.input_method()
        GCDcalculation.calculation_method()
        GCDcalculation.output_method()
    
    except pyip.RetryLimitException:
            print("輸入錯誤次數過多，請重新啟動程式。")
    except Exception as e:
            print(f"發生無法預期的錯誤: {e}")

if __name__=="__main__":
     main()
```

### [key Note]:
def calculation_method(self) -> None:\
input x, y\
while (y != 0):\
    temp = y\
    y = x % y\
    x = temp


1. 條件檢查\
while (y != 0): 當 y 不等於 0 時，進入循環。如果 y 為 0，則跳出循環。這是因為當一個數字為 0 時，最大公因數就是另一個數。
2. 儲存 y 的值:\
temp = y: 將當前的 y 值存入 temp，這樣在下一步中可以保留它的值。
3. 計算餘數:\
y = x % y: 使用 % 運算符計算 x 除以 y 的餘數，並將這個餘數賦值給 y。這一步相當於將 y 更新為 x 和 y 的餘數。
4. 更新 x 的值:\
x = temp: 將之前儲存的 y 值（現在在 temp 中）賦值給 x。這樣在下一次循環中，x 變為原來的 y 值。
5. 循環結束:\
當 y 最終變為 0 時，跳出循環。此時 x 便是兩個數的最大公因數。

# 509 最簡分數
請撰寫一程式，讓使用者輸入二個分數，分別是x/y和m/n（其中x、y、m、n皆為正整數），計算這兩個分數的和為p/q，接著將p和q傳遞給名為compute()函式，此函式回傳p和q的最大公因數（Greatest Common Divisor, GCD）。再將p和q各除以其最大公因數，最後輸出的結果必須以最簡分數表示

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip
import math

class IrreducibleFraction:
    def __init__(self) -> None:
        self.x:int=0
        self.y:int=0
        self.m:int=0
        self.n:int=0
        self.p:int=0
        self.q:int=0
        

    def input_method(self) -> None:
        while True:
            self.x=pyip.inputInt("請輸入第一個分數的分子: \n",min=1)
            self.y=pyip.inputInt("請輸入第一個分數的分母: \n",min=1)
            if self.x<self.y:
                break
            else:
                print("條件是:分子<分母,請從新輸入")
                
        while True:          
            self.m=pyip.inputInt("請輸入第二個分數的分子: \n",min=1)
            self.n=pyip.inputInt("請輸入第二個分數的分母: \n",min=1)
            if self.m<self.n:
                break
            else:
                print("條件是:分子<分母,請從新輸入")

    def calculation_method(self) -> None:
        self.p=self.x*self.n+self.m*self.y
        self.q=self.y*self.n
        self.simplify_method()       
    
    def simplify_method(self) ->None:
        gcd_value=math.gcd(self.p,self.q)
        self.p=self.p//gcd_value
        self.q=self.q//gcd_value
              
    def output_method(self) -> None:
        print(f"{self.x}/{self.y} + {self.m}/{self.n}={self.p}/{self.q}")
    
def main():
    fraction_calculation=IrreducibleFraction()

    try:
        fraction_calculation.input_method()
        fraction_calculation.calculation_method()
        fraction_calculation.output_method()
    
    except pyip.RetryLimitException:
            print("輸入錯誤次數過多，請重新啟動程式。")
    except Exception as e:
            print(f"發生無法預期的錯誤: {e}")

if __name__=="__main__":
     main()
```

# 510 費氏數列
設計說明：
請撰寫一程式，計算費氏數列（Fibonacci numbers），使用者輸入一正整數num (num>=2)，並將它傳遞給名為compute()的函式，此函式將輸出費氏數列前num個的數值。
提示：費氏數列的某一項數字是其前兩項的和，而且第0項為0，第一項為1，表示方式如下：
𝐹0=0\
𝐹1=1\
Fn=Fn-1+Fn-2

##### 程式II-1: 物件導向程式設計（Object-oriented programming，OOP)

```python
import pyinputplus as pyip

class FibonacciNumbers:
    def __init__(self,n:int) -> None:
        self.n=n
        self._current_number=[]
        
    @property
    def current_number(self):
        return self._current_number
    

    def calculation_method(self) -> None:
        if self.n==0:
            self._current_number=[0]
        elif self.n==1:
            self._current_number=[0]
        else:
            self._current_number=[0,1]
            for _ in range(2,self.n):
                next_number=self._current_number[-1]+self._current_number[-2]
                self._current_number.append(next_number)
                           

    def output_method(self) -> None:
        FIB_sequence = " ".join(map(str,self._current_number))
        print(f"費氏數列前{self.n}個的數值為 {FIB_sequence}")

def input_function():
    n=pyip.inputInt("請輸入一個正整數: ")
    return n
  
def main():

    while True:
        try:
            selection: str=pyip.inputYesNo(prompt="Python 510 費氏數列, 繼續計算請按'y'; 離開程式請按'q' ",yesVal="y",noVal="q")
            print()

            if selection=='y':
                n=input_function()
                FIB_calculation=FibonacciNumbers(n)
                FIB_calculation.calculation_method()
                FIB_calculation.output_method()

            else:
                print("程式結束,謝謝使用! ")
                break

            
        except pyip.RetryLimitException as error:
            print(f"輸入超過限制次數: {error}")
            break
        except ValueError as error:
            print(f"輸入值錯誤: {error}")
            break
        except Exception as error:
            print(f"發生無法預期的錯誤: {error}")
            break

if __name__=="__main__":
    main()
```
